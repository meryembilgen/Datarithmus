## Introduction

**What is Exploratory Data Analysis ?**

- Exploratory Data Analysis (EDA) is the process of analyzing and visualizing datasets to uncover patterns, trends, and relationships within the data before applying statistical modeling. Its primary goals include understanding the data structure, identifying anomalies and missing values, and generating hypotheses for further analysis. Common techniques used in EDA include descriptive statistics, data visualization (such as histograms and scatter plots), and correlation analysis. EDA is essential for gaining insights and guiding data-driven decisions in any analysis project.

**How to perform Exploratory Data Analysis ?**

- To perform Exploratory Data Analysis (EDA), start by importing necessary libraries (like Pandas and Matplotlib) and loading your dataset. Understand the data structure using methods like .head(), .info(), and .describe(). Check for missing values with .isnull().sum(), and clean the data by addressing duplicates or inconsistencies. Use visualizations, such as histograms, scatter plots, and box plots, to explore distributions and relationships between variables. Finally, summarize your findings to generate insights that can guide further analysis.

## 10. Imputing Missing Values
- Handle missing values thoughtfully: use mean, median, mode, or more complex imputation methods like forward fill, backward fill, or machine learning models for imputation.

```python
df[["Engine_HP", "Engine_Cylinders", "Make", "Model", "Year"]]
```

```python
df['Engine_HP'] = df.groupby(['Make', 'Model', 'Year'])['Engine_HP'].transform(lambda x: x.fillna(x.mean())) # hepsini doldurmadı
```

```python
df['Engine_Cylinders'] = df.groupby(['Make', 'Model', 'Year'])['Engine_Cylinders'].transform(lambda x: x.fillna(x.mean())) # hepsini doldurmadı
```

```python
df['Number_of_Doors'] = df.groupby(['Make', 'Model', 'Year'])['Number_of_Doors'].transform(lambda x: x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
for group1 in list(df['Make'].unique()):
        for group2 in list(df['List_Price'].unique()):
            cond1 = df['Make']==group1
            cond2 = (df['Make']==group1) & (df['List_Price']==group2)
            mode1 = list(df[cond1]['Engine_Cylinders'].mode())
            mode2 = list(df[cond2]['Engine_Cylinders'].mode())
            if mode2 != []:
                df.loc[cond2, 'Engine_Cylinders'] = df.loc[cond2, 'Engine_Cylinders'].fillna(df[cond2]['Engine_Cylinders'].mode()[0])
            elif mode1 != []:
                df.loc[cond2, 'Engine_Cylinders'] = df.loc[cond2, 'Engine_Cylinders'].fillna(df[cond1]['Engine_Cylinders'].mode()[0])
            else:
                df.loc[cond2, 'Engine_Cylinders'] = df.loc[cond2, 'Engine_Cylinders'].fillna(df['Engine_Cylinders'].mode()[0])
```

```python
for group1 in list(df['Make'].unique()):
        for group2 in list(df['List_Price'].unique()):
            cond1 = df['Make']==group1
            cond2 = (df['Make']==group1) & (df['List_Price']==group2)
            mode1 = list(df[cond1]['Engine_HP'].mode())
            mode2 = list(df[cond2]['Engine_HP'].mode())
            if mode2 != []:
                df.loc[cond2, 'Engine_HP'] = df.loc[cond2, 'Engine_HP'].fillna(df[cond2]['Engine_HP'].mode()[0])
            elif mode1 != []:
                df.loc[cond2, 'Engine_HP'] = df.loc[cond2, 'Engine_HP'].fillna(df[cond1]['Engine_HP'].mode()[0])
            else:
                df.loc[cond2, 'Engine_HP'] = df.loc[cond2, 'Engine_HP'].fillna(df['Engine_HP'].mode()[0])
```

## 13. Outlier Handling 
- Address outliers using methods like capping (clipping to upper/lower bounds), removing, or binning values. Consider keeping outliers if they represent meaningful data. Otherwise dropping the missing or null values.

```python
Q1 = df["List_Price"].quantile(0.25) 
Q3 = df["List_Price"].quantile(0.75)  
IQR = Q3 - Q1  

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Address outliers
outliers_ = df[(df["List_Price"] < lower_bound) | (df["List_Price"] > upper_bound)]
print("Outliers:\n", outliers_)
```
